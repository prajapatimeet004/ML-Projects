```python
import pandas as pd
data = pd.read_csv(r"C:\Users\Admin\OneDrive\Desktop\python\ML\assigment\whitewines.csv")
data
```

```python
#dividing into dependent variable and independet variable
x = data.drop(columns = 'quality')
y = data.quality

#checking for balanced data
for col in data.columns:
    print(data[col].value_counts())
```

```python
#spliting data into training and testing 
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.30,random_state = 42)
```

```python
#model building 
from sklearn.tree import DecisionTreeRegressor
DTR = DecisionTreeRegressor(random_state=42)
#model.fit(X_train, y_train)
DTR.fit(x_train,y_train)
y_pred = DTR.predict(x_test)
```

```python
DTR.score(x_test,y_test)
```

```python
#using cross validation
from sklearn.model_selection import cross_val_score

cross_val_score(DTR,x,y,cv=5).mean()
```

```python
#using shuffle split

from sklearn.model_selection import ShuffleSplit
LOO = ShuffleSplit()
cross_val_score(DTR,x,y,cv=LOO).mean()
```
