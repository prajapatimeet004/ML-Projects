```python
import pandas as pd
data = pd.read_csv(r"C:\Users\Admin\OneDrive\Desktop\python\ML\assigment\mushrooms.csv")
data.columns
```

```python
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
cols = ['type','cap_shape', 'cap_surface', 'cap_color', 'bruises', 'odor',
       'gill_attachment', 'gill_spacing', 'gill_size', 'gill_color',
       'stalk_shape', 'stalk_root', 'stalk_surface_above_ring',
       'stalk_surface_below_ring', 'stalk_color_above_ring',
       'stalk_color_below_ring', 'veil_type', 'veil_color', 'ring_number',
       'ring_type', 'spore_print_color', 'population', 'habitat']
for col in cols:
    data[col] = LE.fit_transform(data[col])
data
```

```python
#checking for unbalanced data
for col in cols:
    print(data[col].value_counts())
```

```python
#assgining x and y (target varaible and independent columns)
x = data.drop(columns = 'type',axis = 1)
y = data.type
```

```python
#spliting data in training and testing part 
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.30,random_state = 42)
```

```python
#importing model and applying model on data
from sklearn.tree import DecisionTreeClassifier
```

```python
DTC = DecisionTreeClassifier(criterion='gini',max_depth = 5,random_state = 42)
DTC.fit(x,y)
y_pred = DTC.predict(x_test)
```

```python
#making report
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
#using cross validation 
from sklearn.model_selection import cross_val_score
cross_val_score(DTC,x,y,cv = 5).mean()
```

```python
#using leave one out method
from sklearn.model_selection import LeaveOneOut
LOO = LeaveOneOut()
cross_val_score(DTC,x,y,cv = LOO).mean()
```

```python
#using shuffle shift
from sklearn.model_selection import ShuffleSplit
SS = ShuffleSplit()
cross_val_score(DTC,x,y,cv = SS).mean()
```
